```r
x<-c(0,50,100,150,200,250,300)
y<-c(8,13115,19780,22612,20083,12430,4)
```

```r
plot(x,y,col="blue", main="Amount Spent on Advertising Bowling Bowls vs. Number Sold",xlab="Amount Spent (thousands of dollars", ylab="Number of Bowling Balls Sold")
```

```r
relation<-lm(y~x)
print(relation)
summary(relation)
```

```r
x_seq <- seq(from=0,to=300, by=0.1)
new_relation<-list(0)
for (i in 1:length(x_seq)) {
    x_new <- data.frame(x=x_seq[i])
    new_relation[i]<-predict(relation, x_new)
}
plot(x,y,col="blue", main="Linear Function",lines(x_seq, new_relation),xlab="x",ylab="y")
```

```r
confint(relation, level=0.95)
```

```r
plot(fitted(relation), residuals(relation))
```
